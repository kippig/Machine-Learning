```python
# After the above file is made, next time it can be loaded directly:

locations = pd.read_csv("airport_locations.csv", index_col="airport")
airports = locations.index.values.tolist()

# (Can use these commands in place of the previous 2 cells)
```

```python
# Create data frame structure to keep track of weather info for current day +/- 1 day:
df = pd.DataFrame(columns=["airport", "date", "tmax", "tmin", "prcp", "snow", "snwd"]) 
# 'date' is date for which forecast is made

# Keep track of the airports with urlopen errors, timeouts or other errors to try again:
failures = []
```

```python
def add_forecast(airport):
    '''Takes as input a 3-letter airport code:
    '''
    try:
        f = urlopen('http://api.wunderground.com/api/{}/yesterday/forecast/q/{},{}.json'.\
                    format(wg_api, locations.loc[airport,'lat'], locations.loc[airport,'long']))
        json_string = f.read()
        parsed_json = json.loads(json_string)

        ystrday = parsed_json["history"]['dailysummary'][0]
        today = parsed_json['forecast']["simpleforecast"]["forecastday"][0]
        tmrrow = parsed_json['forecast']["simpleforecast"]["forecastday"][1]

        date0 = ystrday['date']['year'] + ystrday['date']['mon'] + ystrday['date']['mday']

        date1 = "{}{:0>2d}{:0>2d}".format(
            today['date']['year'], today['date']['month'], today['date']['day'])

        date2 = "{}{:0>2d}{:0>2d}".format(
            tmrrow['date']['year'], tmrrow['date']['month'], tmrrow['date']['day'])

        # Temps need to be in tenths of degrees C
        if ystrday['maxtempm']:
            tmax0 = float(ystrday['maxtempm'])*10
        else:
            tmax0 = -9999
        tmax1 = float(today['high']['celsius'])*10
        tmax2 = float(tmrrow['high']['celsius'])*10

        if ystrday['maxtempm']:
            tmin0 = float(ystrday['mintempm'])*10
        else:
            tmin0 = -9999
        tmin1 = float(today['low']['celsius'])*10
        tmin2 = float(tmrrow['low']['celsius'])*10

        # Units for precipitation are in tenths of mm
        if ystrday['precipm'] == "T":
            prcp0 = 1.0
        else:
            prcp0 = float(ystrday['precipm'])*10    
        prcp1 = float(today['qpf_allday']['mm'])*10
        prcp2 = float(tmrrow['qpf_allday']['mm'])*10

        # Units for snow & snow depth are mm
        if not ystrday['snowfallm']:
            snow0 = 0.0
        elif ystrday['snowfallm'] == "T":
            snow0 = 1.0
        else:
            snow0 = float(ystrday['snowfallm'])*10
        snow1 = float(today['snow_allday']['cm'])*10
        snow2 = float(tmrrow['snow_allday']['cm'])*10

        if not ystrday['snowdepthm']:
            snwd = 0.0
        elif ystrday['snowdepthm'] == "T":
            snwd = 1.0
        else:
            snwd = float(ystrday['snowdepthm'])*10

        df.loc[df.shape[0]] = [airport, date0, tmax0, tmin0, prcp0, snow0, snwd]
        df.loc[df.shape[0]] = [airport, date1, tmax1, tmin1, prcp1, snow1, snwd]
        df.loc[df.shape[0]] = [airport, date2, tmax2, tmin2, prcp2, snow2, snwd]

        f.close()
        time.sleep(15)  # Pauses 15 seconds after each call
    except Exception as e:
        failures.append(airport)
        print "Error for airport {}: {}".format(airport, e)
```

```python
# This should take about 1 hour and 45 mins:

for airport in airports:
    add_forecast(airport)

# Try the airports that didn't work one more time:  
print "\nNow trying again to retrieve weather info where the attempt previously failed.\n"
for i in range(len(failures)):
    add_forecast(failures.pop(0))

# One last time
if failures:
    for i in range(len(failures)):
        add_forecast(failures.pop(0))
```
